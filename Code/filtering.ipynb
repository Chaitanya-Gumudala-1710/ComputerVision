### Loading the Required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/home/chaitanya/computer_vision/images/"
```

##### Loading the panda image

```python
panda_image_path = path + "panda.jpeg" 
panda_image = cv2.imread(panda_image_path, cv2.IMREAD_GRAYSCALE)
# Printing the size of the color image
print('Size of the image is :', panda_image.shape)
plt.figure(figsize=(6, 6))
plt.imshow(panda_image, cmap="gray")
plt.title("Panda image")
plt.show()
```

##### Applying Uniform Weight Filter

```python
# Create a uniform weight kernel (3x3 matrix with all elements set to 0.5)
uniform_weight_kernel = np.full((3, 3), 0.5)
# Display the uniform weight kernel
print(uniform_weight_kernel)
# Apply the uniform weight filter to the panda image 
uniform_weights_filter_image = cv2.filter2D(panda_image, -1, uniform_weight_kernel)
# Plot the original and filtered images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(panda_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(uniform_weights_filter_image, cmap="gray")
plt.title("After applying uniform weight filter")
plt.show()
```

##### Applying Non-Uniform Weight Filter

```python
# Create a non-uniform weight kernel (3x3 matrix with random values between 0 and 1)
non_uniform_weights_kernel = np.random.uniform(0, 1, size=(3, 3))
# Display the non-uniform weight kernel
print(non_uniform_weights_kernel)
# Apply the non-uniform weight filter to the panda image 
non_uniform_weights_filter_image = cv2.filter2D(panda_image, -1, non_uniform_weights_kernel)
# Plot the original and filtered images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(panda_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(non_uniform_weights_filter_image, cmap="gray")
plt.title("Applying non uniform weight filter")
plt.show()
```

##### Applying Gaussian filter

```python
# Set the standard deviation (sigma) for the Gaussian kernel
sigma = 1
# Generate a 1D kernel along the x-axis using the Gaussian function
kernel = np.linspace(-1, 1, 45)
kernel = np.exp(-0.5 * (kernel / sigma) ** 2)
# Normalize the kernel to ensure it sums to 1
kernel /= kernel.sum()
# Create a 2D Gaussian weight kernel using outer product of the 1D kernel
gaussian_weights_kernel = np.outer(kernel, kernel) 
# Display the shape of the Gaussian weight kernel
print(gaussian_weights_kernel.shape)
# Apply the Gaussian weight filter to the panda image
gaussian_weights_filter_image = cv2.filter2D(panda_image, -1, gaussian_weights_kernel)
# Plot the original and filtered images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(panda_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(gaussian_weights_filter_image, cmap="gray")
plt.title("Applying Gaussian weight filter")
plt.show()
```
