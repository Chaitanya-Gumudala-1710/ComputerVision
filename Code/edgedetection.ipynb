### Loading the required Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/home/chaitanya/computer_vision/images/"
```

##### Loading the building image

```python
building_image_path = path + "building.jpeg"
building_image = cv2.imread(building_image_path, cv2.IMREAD_GRAYSCALE)
# Printing the size of the color image
print('Size of the image is :', building_image.shape)
plt.figure(figsize=(6, 6))
plt.imshow(building_image, cmap="gray")
plt.title("Building Image")
plt.show()
```

##### Applying sobel kernel for edge detection

```python
sobel_kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
```

```python
sobel_gradient_x = cv2.filter2D(building_image, -1, sobel_kernel_x)
sobel_gradient_y = cv2.filter2D(building_image, -1, sobel_kernel_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(sobel_gradient_x)
plt.title("Sobel kernel in x-direction")
plt.subplot(1, 2, 2)
plt.imshow(sobel_gradient_y)
plt.title("Sobel kernel in y-direction")
plt.show()
```

```python
sobel_magnitude = np.sqrt(sobel_gradient_x ** 2 + sobel_gradient_y ** 2)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(building_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(sobel_magnitude)
plt.title("Image after applying sobel kernel(magnitude)")
plt.show()
```

```python
sobel_direction = np.arctan2(sobel_gradient_x, sobel_gradient_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(building_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(sobel_direction)
plt.title("Image after applying sobel kernel(direction)")
plt.show()
```

##### Applying prewitt kernel for edge detection

```python
prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
```

```python
prewitt_gradient_x = cv2.filter2D(building_image, -1, prewitt_x)
prewitt_gradient_y = cv2.filter2D(building_image, -1, prewitt_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(prewitt_gradient_x)
plt.title("Prewitt kernel in x-direction")
plt.subplot(1, 2, 2)
plt.imshow(prewitt_gradient_y)
plt.title("Prewitt kernel in y-direction")
plt.show()
```

```python
prewitt_magnitude = np.sqrt(prewitt_gradient_x ** 2 + prewitt_gradient_y ** 2)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(building_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(prewitt_magnitude)
plt.title("Image after applying prewitt kernel(magnitude)")
plt.show()
```

```python
prewitt_direction = np.arctan2(prewitt_gradient_x, prewitt_gradient_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(building_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(prewitt_direction)
plt.title("Image after applying prewitt kernel(direction)")
plt.show()
```

##### Applying Roberts kernel for edge detection

```python
roberts_x = np.array([[0, 1], [-1, 0]])
roberts_y = np.array([[1, 0], [0, -1]])
```

```python
roberts_gradient_x = cv2.filter2D(building_image, -1, roberts_x)
roberts_gradient_y = cv2.filter2D(building_image, -1, roberts_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(roberts_gradient_x)
plt.title("Roberts kernel in x-direction")
plt.subplot(1, 2, 2)
plt.imshow(roberts_gradient_y)
plt.title("Roberts kernel in y-direction")
plt.show()
```

```python
roberts_direction = np.arctan2(roberts_gradient_x, roberts_gradient_y)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(building_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(roberts_direction)
plt.title("Image after applying robert kernel(direction)")
plt.show()
```
