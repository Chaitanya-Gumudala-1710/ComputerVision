### Loading the Required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/home/chaitanya/computer_vision/images/"
```

##### Load, display grayscale images

```python
dog_image_path = path + "dog.jpg"
# Reading the dog image in grayscale 
dog_image = cv2.imread(dog_image_path, cv2.IMREAD_GRAYSCALE)
# Printing the size of the image
print('Size of the image is :', dog_image.shape)
# Displaying the image 
plt.figure(figsize=(6, 6))
plt.imshow(dog_image, cmap='gray')
plt.show()
```

##### 3D plot of grayscale image

```python
# Creating meshgrid for X and Y coordinates based on the shape of the dog image
X, Y = np.meshgrid(range(dog_image.shape[1]), range(dog_image.shape[0]))
# Plotting the 3D surface using X, Y, and the intensity values of the dog image
figure = plt.figure(figsize=(6, 6))
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(X, Y, dog_image, cmap='gray')
plt.show()
```

##### Slicing and displaying the region of interest

```python
# Extracting a region of interest (nose) from the dog image
nose = dog_image[150:225, 150:225]
# Plotting the extracted nose region
plt.figure(figsize=(6, 6))
plt.imshow(nose, cmap='gray')
plt.show()
```

##### Load and Display same image in color and grayscale

```python
panda_image_path = path + "panda.jpeg" 
# Reading the panda image in color 
panda_image = cv2.imread(panda_image_path)
# Printing the size of the color image
print('Size of the image is :', panda_image.shape)
# Reading the panda image in grayscale
panda_image_greyscale = cv2.imread(panda_image_path, cv2.IMREAD_GRAYSCALE)
# Printing the size of the grayscale image
print('Size of the grey scale image is :', panda_image_greyscale.shape)
# Plotting the color and grayscale images side by side
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(panda_image)
plt.title("Color Image")
plt.subplot(1, 2, 2)
plt.imshow(panda_image_greyscale, cmap="gray")
plt.title("Gray scale Image")
plt.show()
```

##### Plot color channel histograms

```python
# Extracting individual color channels from the panda image
red_channel = panda_image[:, :, 0]
green_channel = panda_image[:, :, 1]
blue_channel = panda_image[:, :, 2]

# Calculating histograms for each color channel
red_histogram, red_bins = np.histogram(red_channel.flatten(), bins=256, range=[0,256])
green_histogram, green_bins = np.histogram(green_channel.flatten(), bins=256, range=[0,256])
blue_histogram, blue_bins = np.histogram(blue_channel.flatten(), bins=256, range=[0,256])

# Plotting the color histograms
plt.figure(figsize=(16, 8))
plt.plot(red_bins[:-1], red_histogram, color="red", label="Red Channel")
plt.plot(green_bins[:-1], green_histogram, color="blue", label="Green Channel")
plt.plot(blue_bins[:-1], blue_histogram, color="green", label="Blue Channel")
plt.legend()
plt.show()
```

##### Visualize 3D RGB channels

```python
# Creating meshgrid for X and Y coordinates based on the shape of the panda image
X, Y = np.meshgrid(range(panda_image.shape[1]), range(panda_image.shape[0]))

# Plotting the 3D surfaces for each RGB channel
figure = plt.figure(figsize=(10, 10))
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(X, Y, red_channel, color="red")
axis.plot_surface(X, Y, green_channel, color="green")
axis.plot_surface(X, Y, blue_channel, color="blue")
axis.set_xlabel('X')
axis.set_ylabel('Y')
axis.set_zlabel('Pixel Value')
axis.set_title('RGB Channels')
plt.show()
```

##### Loading and Displaying grayscale images of elephant and mountains

```python
# Reading the elephant and mountain images in grayscale 
elephant_image_path = path + "elephant.jpg"
elephant_image = cv2.imread(elephant_image_path, cv2.IMREAD_GRAYSCALE)
print("Size of Elephant image is :", elephant_image.shape)
mountain_image_path = path + "mountain.jpg"
mountain_image = cv2.imread(mountain_image_path, cv2.IMREAD_GRAYSCALE)
print("Size of Mountain image is :", mountain_image.shape)
# Plotting the images side by side
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(elephant_image, cmap="gray")
plt.title("Elephant Image")
plt.subplot(1, 2, 2)
plt.imshow(mountain_image, cmap="gray")
plt.title("Mountains Image")
plt.show()
```

##### Displaying combined grayscale image of elephant and mountain

```python
# Combining the two grayscale images
combined_image = elephant_image + mountain_image
# Plotting the combined image
plt.figure(figsize=(6, 6))
plt.imshow(combined_image, cmap="gray")
plt.title("Combined Image")
plt.show()
```

##### Loading and Displaying image of original and scaled tiger.

```python
tiger_image_path = path + "tiger.jpg"
# Reading the tiger image in grayscale
tiger_image = cv2.imread(tiger_image_path, cv2.IMREAD_GRAYSCALE)
# Scaling the intensity values of the tiger image by a scalar value (3 in this case)
scaled_tiger_image = 3 * tiger_image
print("Size of Tiger image is :", tiger_image.shape)
# Plotting the images side by side
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(tiger_image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(scaled_tiger_image, cmap="gray")
plt.title("Image multiplied by scalar value")
plt.show()
```

##### Displaying subtracted grayscale image of elephant and mountain

```python
# Subtracting the two grayscale images
subtracted_image = elephant_image - mountain_image
# Plotting the combined image
plt.figure(figsize=(6, 6))
plt.imshow(subtracted_image, cmap="gray")
plt.title("Subtracted Image")
plt.show()
```

##### Adding gaussian blur to the panda image

```python

blurred_panda_image = cv2.GaussianBlur(panda_image, (25, 25), 5)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(panda_image)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(blurred_panda_image)
plt.title("Blurred Panda Image")
plt.show()
```
